```python
import xmltodict
import pprint
```

```python
data = {}

for elem in dom.getElementsByTagName('page-layout'):
    
    for child1 in elem.childNodes:
        if len(child1.childNodes) > 0:
            
            node_name       = unicodedata.normalize('NFKD', child1.nodeName).encode('ascii','ignore')
            
            if node_name == "page-margins":
                
                for child2 in child1.childNodes:
                    
                    if len(child2.childNodes) > 0:
                        
                        node_name       = unicodedata.normalize('NFKD', child2.nodeName).encode('ascii','ignore')
                        data[node_name] = unicodedata.normalize('NFKD', child2.childNodes[0].data).encode('ascii','ignore')
            
            else:
            
                data[node_name] = unicodedata.normalize('NFKD', child1.childNodes[0].data).encode('ascii','ignore')

data["lines"] = []

cur_line = None

for elem in dom.getElementsByTagName('measure'):
    
    if (elem.nodeType != xml.dom.minidom.Node.ELEMENT_NODE):
        continue
    
    curr_measure = {}
    
    curr_measure["number"] = unicodedata.normalize('NFKD', elem.getAttribute("number")).encode('ascii','ignore')
    curr_measure["width"] = unicodedata.normalize('NFKD', elem.getAttribute("width")).encode('ascii','ignore')
    
    for print1 in elem.getElementsByTagName('print'):
        
        #print print1.getAttribute("new-system") == "", 1
        if (print1.getAttribute("new-system") == "" or print1.getAttribute("new-system") == "yes"):
            
            if cur_line != None:
                data["lines"].append(cur_line)
            
            cur_line = {}
            cur_line["print"] = {}
            cur_line["measures"] = []
            cur_line["print"] = xmltodict.parse(print1.toxml())["print"]
        
        
    
    cur_line["measures"].append(curr_measure)
    

data["lines"].append(cur_line)
    
data

```
