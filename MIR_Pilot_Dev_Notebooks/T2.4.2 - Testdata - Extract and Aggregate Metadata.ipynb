```python
elements_to_extract = ["title",
                       "about"
                       "aggregations.0.about", 
                       "aggregations.0.edmDataProvider.def", 
                       "aggregations.0.edmProvider.def",
                       "aggregations.0.edmRights.def",
                       "aggregations.0.webResources.0.about", 
                       "aggregations.0.edmIsShownBy", 
                       "proxies.0.dctermsCreated.def", 
                       "proxies.0.dctermsExtent.def", 
                       "proxies.0.dctermsMedium.def", 
                       "proxies.0.dctermsSpatial.def",
                       "proxies.0.dctermsTOC.def",
                       "proxies.0.dctermsIsPartOf.def",
                       "proxies.0.dcFormat.def",
                       "proxies.0.dcCoverage.def",
                       "proxies.0.dcContributor.def",
                       "proxies.0.dcCreator.def",
                       "proxies.0.dcSubject.def",                       
                       "proxies.0.dcLanguage.def",
                       "proxies.0.dcDescription.def",
                       "proxies.0.dcType.def",
                       "proxies.0.dcSource.def",
                       "proxies.0.dcRights.def",
                       "europeanaAggregation.edmCountry.def",
                       "europeanaAggregation.edmLanguage.def",
                       "europeanaCollectionName"
                      ]

aggregations_elements_to_extract = ["Provider", "Meidum", "Spatial", "TOC", "title", "edmDatasetName", "europeanaCollectionName", "link", "year", "id",
                       "guid", "dataProvider", "language"]

proxy_elements_to_extract = ["IsPartOf", "Meidum", "Spatial", "TOC"]


def extract_instance(data):
    instance = []

    object_data = data["object"]
    
    # extract top-level elements
    for elem_desc in elements_to_extract:

        try:
            dat = data["object"]
            
            for elem in elem_desc.split("."):
                
                #print ">", elem
                
                if elem.isdigit():
                    dat = dat[int(elem)]
                elif elem in dat.keys():
                    dat = dat[elem]
                else:
                    raise Exception("not an element '%s'" % (elem))
          
            if type(dat) == list:
                instance.append(', '.join(dat))
            else:
                instance.append(dat)
          
        except Exception as e:
            instance.append(None)
            #print "exception", e
            
    return instance

#extract_instance(data)
```

```python
extracted_data = []

idx = 0

failed = 0

for subdir in glob.glob(os.path.join(METADATA_PATH, "*")):

    if not os.path.isdir(subdir):
        continue
                
    for json_file_path in glob.glob(os.path.join(subdir, "*.json")):
        
        tmp = json_file_path.replace("\\","/").replace(".json","").split("/")
        
        eus_id = "/%s/%s" % (tmp[-2],tmp[-1])
                
        try:
            with open(json_file_path) as df:
                data = json.load(df)
            
            instance_data = [eus_id]
            instance_data.extend(extract_instance(data))
            
            extracted_data.append(instance_data)
            
        except Exception as e:
            #print "***", json_file_path, e
            failed += 1
                        
        idx += 1
        
        if idx % 10000 == 0:
            print idx
 
        #if idx > 10:
        #    break
        
    #break
```
