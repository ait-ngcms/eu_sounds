### Annotations

```python
data = ""

page_nr = 0

for page in page_order:
    
    f_name = "{0}/{1}.txt".format(layout_desc_path, page)
    
    if os.path.exists(f_name):
    
        page_data = ""
    
        #print page_nr, f_name
        
        csv = pd.read_csv(f_name, index_col=None)
        
        for i in range(csv.shape[0]):
            line = csv.iloc[i].values[0]
            x = csv.iloc[i].values[1]
            y = csv.iloc[i].values[2]
            width = csv.iloc[i].values[3]
            height = csv.iloc[i].values[4]
            
            page_height = img_sizes[page_nr][1]
            page_width  = img_sizes[page_nr][0]
            
            #print line, x, y, width, height, page_height, page_width
            
            #print line, float(x)/page_height, float(y)/page_width, float(width)/page_width, float(height)/page_height
            
            
            page_data = "{0},[{1},{2},{3},{4}]".format(page_data, 
                                                       float(x)/page_width, 
                                                       float(y)/page_height, 
                                                       float(width)/page_width, 
                                                       float(height)/page_height)
            
            
            
            #break
        
        data = "{0}, [{1}]".format(data,page_data[1:])
        
    else:
        
        #print page_nr, []
        data = "{0}, []".format(data)
        
    
    page_nr += 1
    
data = "[{0}]".format(data[2:])

print data
```
